```python
import gc # to manage memory and garbage collection

class Myobject:
    def __init__(self, name):
        self.name = name
        print(f"Object {self.name} created")

    def __del__(self):
        print(f"Object {self.name} destroyed")

# Create circular references
obj1 = Myobject("obj1")
obj2 = Myobject("obj2")         
obj1.ref = obj2
obj2.ref = obj1

# Delete references
del obj1
del obj2

# Manually trigger garbage collection
gc.collect()    
```
