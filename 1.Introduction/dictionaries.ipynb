```python
empty_dict = {}

print(type(empty_dict))  # Output: <class 'dict'>
```

```python
empty_dict = dict()
print(type(empty_dict))  # Output: <class 'dict'>
```

```python
student = {
    "name": "John Doe",
    "age": 21,
    "courses": ["Math", "CompSci"]} # Key should be unique

print(student)
```

```python
# Accessing dictionary elements
print(student["name"])  # Output: John Doe
print(student.get("age"))  # Output: 21 
print(student.get("address", "Not Found"))  # Output: Not Found
```

```python
# Modifying dictionary elements
student["age"] = 22
student["address"] = "123 Main St"
print(student)
```

```python
del student["courses"]
print(student)
```

```python
print(student.keys())
```

```python
print(student.values())
```

```python
print(student.items())
```

```python
# Shallow Copy vs Deep Copy
original_dict = {'a': 1, 'b': 2, 'c': [3, 4, 5]}
shallow_copy = original_dict.copy()
deep_copy = dict(original_dict) 
deep_copy['c'] = original_dict['c'][:]
shallow_copy['c'][0] = 99
print("Original Dictionary:", original_dict)  # 'c' will reflect change
print("Shallow Copy:", shallow_copy)  # 'c' will reflect change
print("Deep Copy:", deep_copy)  # 'c' will not reflect change
```

```python
# Nested Dictionaries
nested_dict = {
    "student1": {
        "name": "Alice",
        "age": 23
    },
    "student2": {
        "name": "Bob",      
        "age": 24
    }
}   
print(nested_dict["student1"]["name"])  # Output: Alice
```

```python
# iterating through a nested dictionary
for student_id, details in nested_dict.items():
    print(f"{student_id}:")
    for key, value in details.items():
        print(f"  {key}: {value}")  
```

```python
# Dictionary Comprehensions
squares = {x: x*x for x in range(6)}
print(squares)  # Output: {0: 0, 1: 1, 2: 4, 3: 9, 4: 16, 5: 25}
```

```python
# conditional dictionary comprehension
even_squares = {x: x*x for x in range(6) if x % 2 == 0}
print(even_squares)  # Output: {0: 0, 2:
```

```python
# Frquency Count using Dictionary Comprehension
numbers = [1, 2, 2, 3, 3, 3, 4]
frequency = {num: numbers.count(num) for num in set(numbers)}
print(frequency)  # Output: {1: 1, 2: 2, 3: 3, 4: 1} 4: 16}
```

```python
# merge two dictionaries
dict1 = {'a': 1, 'b': 2}
dict2 = {'b': 3, 'c': 4}    
merged_dict = {**dict1, **dict2} #** is used to unpack the dictionaries
print(merged_dict)  # Output: {'a': 1, 'b': 3, 'c': 4}
```
