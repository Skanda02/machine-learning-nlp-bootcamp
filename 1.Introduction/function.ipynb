```python
# Positional arguments
def print_values(*args):
    for value in args:
        print(value)

print_values(1, 2, 3, 'hello')
```

```python
# Keyword arguments
def print_info(**kwargs):
    for key, value in kwargs.items():
        print(f"{key}: {value}")

print_info(name="Alice", age=30, city="New York")
```

```python
# positional and keyword arguments together
def display_info(*args, **kwargs):  
    for value in args:
        print(value)
    for key, value in kwargs.items():
        print(f"{key}: {value}")

display_info(10, 20, name="Bob", country="USA") #positional arguments first then keyword arguments
```

```python
# return multiple values from a function
def calculate(a, b):
    sum_result = a + b
    product_result = a * b
    return sum_result, product_result   

sum_val, prod_val = calculate(5, 10)
print(f"Sum: {sum_val}, Product: {prod_val}")
```

# Lambda function
Lambda fuctions are small anonymous functions defined using the lambda keyword. They can take any number of arguments but can only have a single expression.

```python
lanbda_add = lambda x, y: x + y


result = lanbda_add(3, 7)
print(f"Lambda Addition Result: {result}")
```

```python
numbers = [1, 2, 2, 3, 3, 3, 4]
mapped_numbers = list(map(lambda x: x * 2, numbers)) # What map does is it applies the lambda function to each element in numbers
                                                    # if it is not converted to list it will return map object
print(f"Mapped Numbers: {mapped_numbers}")
```

# Map Function

```python
number1 = [1, 2, 3] 
number2 = [4, 5, 6]

added_numbers = list(map(lambda x, y: x + y, number1, number2))
print(f"Added Numbers: {added_numbers}")
```

```python
# Filter function example
filtered_numbers = list(filter(lambda x: x % 2 == 0, numbers)) #filter returns only those elements which satisfy the condition
print(f"Filtered Even Numbers: {filtered_numbers}") 
```
